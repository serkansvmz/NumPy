```python
# z- score
```

```python
data = np.array ([10,12,37,25,100,120])
```

```python
data
```

```python
mean = np.mean(data)
```

```python
mean
```

```python
data.sum() / len(data)
```

```python
std = np.std(data)
```

```python
std
```

```python
z_score = (data-mean) / std
```

```python
z_score
```

```python
data[z_score > 1]
```
